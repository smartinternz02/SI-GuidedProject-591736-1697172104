```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("/content/drive/MyDrive/Dataset/Adaptability_Level.xlsx")
df
```

```python
df.describe()
```

```python
df.drop('Age',axis=1,inplace=True)
```

```python
df.drop('Class Duration',axis=1,inplace=True)
```

```python
df.drop('Location',axis=1,inplace=True)
```

```python
df.drop('Load-shedding',axis=1,inplace=True)
```

```python
df.drop('Self Lms',axis=1,inplace=True)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
ml=LabelEncoder()
df["Gender"]=ml.fit_transform(df["Gender"])
df["Education Level"] = ml.fit_transform(df [ "Education Level"])
df["Institution Type"] = ml.fit_transform(df [ "Institution Type"])
df["IT Student"] = ml.fit_transform(df["IT Student"])
df["Financial Condition"] = ml.fit_transform(df["Financial Condition"])
df["Internet Type"] = ml.fit_transform(df["Internet Type"])
df["Network Type"] = ml.fit_transform (df["Network Type"])
df["Device"] = ml.fit_transform(df["Device"])
df["Adaptivity Level"] = ml.fit_transform(df [ "Adaptivity Level"])
```

```python
x = df.drop(['Adaptivity Level'],axis=1)
y = df['Adaptivity Level']
```

```python
x.info()
```

```python
y.info()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2, random_state=42)
X_train.shape,X_test.shape
```

```python
#Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
from sklearn.metrics import confusion_matrix
predictions_2 = rfc.predict(X_test)
pred_train2=rfc.predict(X_train)
confusion_matrix(predictions_2, Y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report (Y_test, predictions_2))
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy_test: ", accuracy_score (predictions_2, Y_test))
print("Accuracy_train : ", accuracy_score (pred_train2, Y_train))
```

```python
#Decision Tree Classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
model1 = DecisionTreeClassifier (random_state=42)
model1.fit(X_train, Y_train)
```
